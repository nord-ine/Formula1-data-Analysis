Le dataset ne nous fournit pas la puissance de la voiture, cette donne n'tant pas quantifiable (mis  part peut-tre la puissance du moteur). Nous devons donc essayer nous-mme de quantifier ce paramtre.

Dans le paddock, les deux ou trois curies dominatrices arrivent  raliser des tours souvent plus d'une seconde plus rapide que les curies de milieu de plateau et les <i>backmarkers</i> : on cherche donc un coefficient $\mu_x$ permettant de modliser cette disparit. Nous choisirons $\mu_x = \frac{1}{tan(0.1x)^{0.7}}$, avec $x$ la position au championnat constructeurs de l'anne prcdente nous appellerons cette feature **"constructorPower"**.

Bien entendu, les avances technologiques d'une curie ne se conservent pas toujours d'une anne sur l'autre : un exemple flagrant en est l'curie Brawn GP, victorieuse en 2009 grce  un systme  la lgalit douteuse mais seulement 4e en 2010, sous l'gide de son acqureur Mercedes. Ainsi, on utilise un deuxime coefficient, que nous appellerons coefficient de forme **"constructorForm"**, qui reprend l'quation du coefficient de puissance en remplacant $x$ par la position actuelle au championnat constructeur : ce coefficient progresse donc au fur et  mesure de la saison.

La performance d'un pilote est galement,  premire vue, une donne trs importante pour bien russir la prdiction. Similairement au cas des coefficients de forme et de performance, nous ne disposons pas d'une colonne dans le dataset nous indiquant directement le niveau d'un pilote. Nous allons ainsi extraire cette information avec les donnes dont on dispose : <br>
Pour un pilote $i$, on cre le coefficient de performance $\mu_i$ de la manire suivante : $\mu_i = \frac{1}{1 + \exp^{((x_j - x_i) + (r_j - r_i))}}$ avec

- $x_i$ la position moyenne de qualifications du pilote
- $x_j$ la position moyenne de qualifications de son quipier
- $r_i$ le rsultat de course moyen du pilote
- $r_j$ le rsultat de course moyen de son quipier


```python
def getTeammate(raceId,driverId):
    result = results[results['raceId'] == raceId]
    constructor = result[result['driverId'] == driverId]['constructorId'].iloc[0]
    current_round = races['round'].loc[raceId]
    try:
        teammate = result[(results['constructorId'] == constructor) & (results['driverId'] != driverId)]['driverId'].iloc[0]
    except Exception:
        if current_round > 1: #Si ce n'est pas la premire course de la saison, on retourne l'quipier de la course prcdente
            return getTeammate(raceId-1, driverId)
        else:
            return getTeammate(raceId+1, driverId)
    return teammate
    
def getAverageQualifyingPosition(raceId, driverId):
    '''
    In : raceId (int)
         driverId (int)
    Out : Le rsultat moyen en qualifications pour cette saison (float)
    '''
    year = races['year'].loc[raceId]
    current_round = races['round'].loc[raceId]
    quali = 0
    nbMissing = 0
    for i in range(1,current_round+1):
        myrace = races[(races['year'] == year) & (races['round'] == i)].index[0]
        try:
            quali += qualifying[(qualifying['raceId'] == myrace) & (qualifying['driverId'] == driverId)]['position'].iloc[0]
        except Exception: #Si un pilote n'a pas particip  une sance de qualifications
            nbMissing +=1
    quali /= 1.0*(current_round-nbMissing)
    return quali

def getAverageFinalPosition(raceId, driverId):
    '''
    In : raceId (int)
         driverId (int)
    Out : Le rsultat moyen en course pour cette saison (float)
    '''
    year = races['year'].loc[raceId]
    current_round = races['round'].loc[raceId]
    pos = 0
    nbMissing = 0
    for i in range(1,current_round+1):
        try:
            myrace = races[(races['year'] == year) & (races['round'] == i)].index[0]
            #print("{}, {}\r".format(myrace, driverId)),
            pos += results[(results['raceId'] == myrace) & (results['driverId'] == driverId)]['positionOrder'].iloc[0]
        except Exception:
            nbMissing +=1
    pos /= 1.0*(current_round-nbMissing)
    return pos

def getDriverName(driverId):
    '''
    In : driverId (int) : numro d'identification du pilote
    Out : nom et prnom du pilote (string)
    '''
    try:
        name = str(drivers['forename'].loc[driverId])+' '+str(drivers['surname'].loc[driverId])
        return name
    except UnicodeDecodeError:
        return driverRef
```

### Cration d'un rseau de neurones avec TensorFlow

Typiquement, TensorFlow permet d'obtenir une bonne prcision sur les rsultats si on sait ce qu'on fait.
Nous allons crer et entraner ce rseau de neurone avec TensorFlow puis comparer ses prdictions avec notre modle lineaire.

```python
def build_model():
  model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model
model = build_model()
```

```python
EPOCHS = 100

history = model.fit(
  std_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=1)
test_predictions = model.predict(std_test_data).flatten()
```

```python
# Optimisation directe
train = train_dataset.to_numpy()
lab = train_labels.to_numpy()
test = test_dataset.to_numpy()
test_lab = test_labels.to_numpy()

def optimize(data,weights,labels,niter=100,speed=0.05,tol=0.00001,i=0):
    MM, MY = np.dot(np.transpose(data), data), np.dot(np.transpose(data), labels)
    return np.dot(np.linalg.inv(MM), MY)
def predict(data, weights):
    return np.dot(data,weights)
myweights = optimize(train, np.random.rand(train.shape[1]), lab)
mypreds = predict(test, myweights)
```
